```python
url = "https://api.fda.gov/food/enforcement.json"
```

```python
dfs = [] # collects dataframes

# Iterating over the range to fetch 25,000 records in batches of 1,000
for i in range(25):
    params = {
        "limit": 1000,
        "skip": i * 1000
    }

    # Fetching data from the API
    data = requests.get(url, params=params).json()

    # Append dataframe of 1000 rows to list of dfs
    dfs.append(pd.DataFrame(data["results"])) 

# Combine dataframes
df = pd.concat(dfs, ignore_index=True)
```

```python
df["classification"] = df["classification"].replace({
    "Class I": "1",
    "Class II": "2",
    "Class III": "3"
})
df["classification"] = df["classification"].astype("Int64")
```

```python
duration_df = df.dropna(subset=["duration"])
```

```python
duration_df = duration_df.sort_values("duration")
```

```python
duration_df["duration"].head()
```

```python
duration_df["duration"].describe()
```

```python
import matplotlib.pyplot as plt
```

```python
ax = duration_df["duration"].dt.days.hist(
    bins=20,
    color="#3498db",
    edgecolor="white",
    alpha=0.8
)

plt.title("Distribution of Duration (Days)", fontsize=16, fontweight="bold", pad=20)
plt.xlabel("Duration (Days)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
df['duration_days'] = df['duration'].dt.days
monthly_average = df.groupby(pd.Grouper(key='recall_initiation_date', freq='QE'))['duration_days'].mean()

# Plotting the average duration per quarter
plt.plot(monthly_average.index, monthly_average, marker='.', linestyle='-', color="cornflowerblue")
plt.title('Average Recall Duration Per Quarter')
plt.xlabel('Quarter')
plt.ylabel('Average Duration (days)')
plt.grid()
plt.xticks(rotation=45)
plt.show()
```
