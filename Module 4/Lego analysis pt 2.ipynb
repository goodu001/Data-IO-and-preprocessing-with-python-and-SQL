```python
# Check unique values within each group

query = """
SELECT theme_name, COUNT(DISTINCT theme_id) AS theme_count
FROM sets_with_themes
GROUP BY theme_name
ORDER BY theme_count DESC;
"""
# Execute the query and store the result in a DataFrame
df = pd.read_sql_query(query, connection)
df.head(10)
```

```python
# Explore largest sets in each theme 

# Define the SQL query
query = """
SELECT theme_name, MAX(num_parts)
FROM sets_with_themes
GROUP BY theme_name;
"""
# Execute the query and store the result in a DataFrame
df = pd.read_sql_query(query, connection)
df.head(10)
```

```python
# Define the SQL query
query = """
SELECT *, MAX(num_parts)
FROM sets_with_themes
GROUP BY theme_name;
"""
# Execute the query and store the result in a DataFrame
df = pd.read_sql_query(query, connection)
df.head(10)
```

```python
# Define the SQL query
query = """
SELECT *, MIN(num_parts)
FROM sets_with_themes
GROUP BY theme_name
ORDER BY num_parts;
"""
# Execute the query and store the result in a DataFrame
df = pd.read_sql_query(query, connection)
df.head(10)
```

```python
# Define the SQL query
query = """
SELECT theme_name, COUNT(*) as set_count, MIN(num_parts) AS min_parts, MAX(num_parts) AS max_parts, AVG(num_parts) AS avg_parts
FROM sets_with_themes
GROUP BY theme_name;
"""
# Execute the query and store the result in a DataFrame
df = pd.read_sql_query(query, connection)
df
```

```python
# Define the SQL query
query = """
SELECT theme_name, COUNT(*) AS set_count
FROM sets_with_themes
GROUP BY theme_name
HAVING COUNT(*) > 50;
"""
# Execute the query and store the result in a DataFrame
df = pd.read_sql_query(query, connection)
df
```

```python
# Filter for themes with 10 sets, then show avg number of parts in each

# Define the SQL query
query = """
SELECT theme_name, COUNT(theme_id) as set_count, AVG(num_parts) AS avg_parts
FROM sets_with_themes
GROUP BY theme_name
HAVING COUNT(theme_id) > 10
ORDER BY avg_parts DESC;
"""
# Execute the query and store the result in a DataFrame
df = pd.read_sql_query(query, connection)
df
```
