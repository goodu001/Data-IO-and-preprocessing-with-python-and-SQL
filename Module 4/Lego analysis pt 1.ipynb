```python
# 1995 and 1996, with theme Aquanauts

# Define the SQL query
query = """
SELECT *
FROM sets_with_themes
WHERE year IN (1995, 1996) AND theme_name = "Aquanauts";
"""
# Execute the query and store the result in a DataFrame
df = pd.read_sql_query(query, connection)
df
```

```python
# Define the SQL query
query = """
SELECT *
FROM sets_with_themes
WHERE theme_name = "Aquanauts" OR theme_name = "Aquasharks"
ORDER BY theme_name;
"""
# Execute the query and store the result in a DataFrame
df = pd.read_sql_query(query, connection)
df
```

```python
# Focus only on larger sets

# Define the SQL query
query = """
SELECT *
FROM sets_with_themes
WHERE theme_name = "Aquanauts" OR theme_name = "Aquasharks" AND num_parts > 100
ORDER BY theme_name;
"""
# Execute the query and store the result in a DataFrame
df = pd.read_sql_query(query, connection)
df
```

```python
# Define the SQL query
query = """
SELECT *
FROM sets_with_themes
WHERE (theme_name = "Aquanauts" OR theme_name = "Aquasharks") AND num_parts > 100
ORDER BY theme_name;
"""
# Execute the query and store the result in a DataFrame
df = pd.read_sql_query(query, connection)
df
```

```python
# Define the SQL query
query = """
SELECT name, num_parts,
CASE
    WHEN num_parts < 10 THEN "Small"
    WHEN num_parts BETWEEN 10 AND 172 THEN "Medium"
    ELSE "Large"
END AS set_size
FROM sets_with_themes;
"""
# Execute the query and store the result in a DataFrame
df = pd.read_sql_query(query, connection)
df
```

```python
df.head()
```

```python
df.groupby("set_size")["num_parts"].mean()
```

```python
# Define the SQL query
query = """
SELECT *
FROM sets_with_themes
WHERE parent_theme_id IS NULL;
"""

# Execute the query and store the result in a DataFrame
df = pd.read_sql_query(query, connection)
df.head(10)
```

```python
# Define the SQL query
query = """
SELECT *
FROM sets_with_themes
WHERE parent_theme_id IS NOT NULL;
"""

# Execute the query and store the result in a DataFrame
df = pd.read_sql_query(query, connection)
df.head(10)
```

```python
# Copy theme id into parent_id column if that column is null

# Define the SQL query
query = """
SELECT theme_id, parent_theme_id,
       COALESCE(parent_theme_id, theme_id) AS updated_parent_theme_id
FROM sets_with_themes;
"""

# Execute the query and store the result in a DataFrame
df = pd.read_sql_query(query, connection)
df.head(10)
```
