```python
import pandas as pd
import sqlite3
```

```python
connection = sqlite3.connect("lego_sets.db")
```

```python
# Define the SQL query
query = """
SELECT *
FROM sets_with_themes;
"""
# Execute the query and store the result in a DataFrame
df = pd.read_sql_query(query, connection)
df
```

```python
# Define the SQL query
query = """
SELECT *
FROM sets_parts;
"""
# Execute the query and store the result in a DataFrame
df = pd.read_sql_query(query, connection)
df
```

```python
# Define the SQL query
query = """
SELECT *
FROM sets_with_themes
LEFT JOIN sets_parts
ON sets_with_themes.set_num = sets_parts.set_num;
"""
# Execute the query and store the result in a DataFrame
df = pd.read_sql_query(query, connection)
df
```

```python
# Define the SQL query
query = """
SELECT *
FROM sets_with_themes
LEFT JOIN sets_parts
ON sets_with_themes.set_num = sets_parts.set_num
WHERE sets_with_themes.set_num = "8480-1";
"""
# Execute the query and store the result in a DataFrame
df = pd.read_sql_query(query, connection)
df
```

```python
# Examine unique parts included in sets in each year
# Define the SQL query
query = """
SELECT year, COUNT(DISTINCT part_num) AS parts_per_year
FROM sets_with_themes
LEFT JOIN sets_parts
ON sets_with_themes.set_num = sets_parts.set_num
GROUP BY year
"""
# Execute the query and store the result in a DataFrame
df = pd.read_sql_query(query, connection)
df
```

```python
import seaborn as sns
sns.lineplot(df["parts_per_year"])
```
